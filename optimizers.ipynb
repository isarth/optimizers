```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import sqrt as sq
```

```python
def y(x):
    return x**2
```

```python
x=35
loss=[]
for j in range(100):
    loss.append(y(x))
    x=x-2*x*0.01
#plt.plot(loss)
```

```python
x=35
v=0
loss_momen=[]
for j in range(100):
    loss_momen.append(y(x))
    v=0.9*v+2*x
    x=x-0.01*v    
```

```python
x=35.0
loss_adagrad=[]
cache=0.0
for n in range(100):
    loss_adagrad.append(y(x))
    cache+=4*x*x 
    x+=-((2*x*0.01)/(sq(cache)+0.0000001))
```

```python
plt.plot(loss_adagrad)
```

```python
x=35.0
loss_rms=[]
cache=0.0
grad=0.0
for n in range(100):
    loss_rms.append(y(x))
    cache=0.99*cache+(0.01)*4*x*x  
    grad=2*x
    x+=-(grad*0.01)/(sq(cache)+0.0000001)
```

```python
x=35.0
loss_adam=[]
m=0.0
v=0.0
for i in range(100):
    loss_adam.append(y(x))
    m=0.9*m+0.1*2*x
    v=0.99*v+0.01*4*x*x
    #m=(m)/(1.0-(0.9**(i+1)))
    #v=(v)/(1.0-(0.99**(i+1)))
    x+=-(0.01*m)/(sq(v)+0.0000001)
 
```

```python
plt.plot(loss_adam)
```

```python
plt.plot(loss_adam)
plt.plot(loss_adagrad)
```

```python
plt.plot(loss)
plt.plot(loss_momen)
plt.plot(loss_adagrad)
plt.plot(loss_rms)
plt.plot(loss_adam,'orange')
plt.legend(('sgd','momentum','adagrad','RMSprop','Adam'))
plt.show()
```

```python
x=35
loss_momen2=[]
v=0.0
for i in range(100):
    x=x-0.01*2*x
    v=v*0.9-0.01*2*x
    x+=v
    loss_momen2.append(y(x))   
```

## Nestrov Momentum

```python

x=35

v_pre=0.0
v=0.0
loss_nestrov=[]
for i in range(100):
    v_pre=v
    v=0.9*v-0.01*2*x
    x+=-0.9*v_pre+(1.9)*v
    loss_nestrov.append(y(x))
```

```python
plt.plot(loss_nestrov)
```
