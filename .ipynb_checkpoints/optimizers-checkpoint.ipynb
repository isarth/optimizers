```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import sqrt as sq
```

```python
def y(x):
    return x**2
```

```python
x=35
loss=[]
for j in range(100):
    loss.append(y(x))
    x=x-2*x*0.01
#plt.plot(loss)
```

```python
x=35
v=0
loss_momen=[]
for j in range(100):
    loss_momen.append(y(x))
    v=0.9*v+2*x
    x=x-0.01*v    
```

```python
x=35.0
loss_adagrad=[]
hist_grad=0.0
grad=0.0
for n in range(100):
    loss_adagrad.append(y(x))
    hist_grad=hist_grad+4*x*x  
    grad=2*x
    x=x-(grad*0.01)/(sq(hist_grad)+0.0000001)
```

```python
x=35.0
loss_rms=[]
hist_grad=0.0
grad=0.0
for n in range(100):
    loss_rms.append(y(x))
    hist_grad=hist_grad*0.99+0.01*4*x*x  
    grad=2*x
    x=x-(grad*0.01)/(sq(hist_grad)+0.0000001)
```

```python
plt.plot(loss)
plt.plot(loss_momen)
plt.plot(loss_adagrad)
plt.plot(loss_rms)
plt.legend(('sgd','momentum','adagrad','RMSprop'))
plt.show()
```
