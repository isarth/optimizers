```python
import optimizers_
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(0,100,(1000))
def f(x):
    return x**2
def df(x):
    return 2*x
```

```python
sgd=optimizers_.SGD(x,f,df,0.01)
for i in range(100):
    sgd.step()
```

```python
mom=optimizers_.momentum(x,f,df,0.01)
for i in range(100):
    mom.step()
```

```python
nmom=optimizers_.nestrov_momentum(x,f,df,0.01)
for i in range(100):
    nmom.step()
```

```python
adagrad=optimizers_.Adagrad(x,f,df,0.01)
for i in range(100):
    adagrad.step()
```

```python
adam=optimizers_.Adam(x,f,df,0.01)
for i in range(100):
    adam.step()
```

```python
sgd.plot()
mom.plot()
nmom.plot()
adagrad.plot()
adam.plot()
plt.legend(('SGD','Momentum','Nestrov Mom','Adagrad','Adam'))
plt.savefig("optimizers.png")
plt.show()
```
